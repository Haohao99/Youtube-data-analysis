```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
from wordcloud import WordCloud, STOPWORDS
%matplotlib inline
```

# Word Cloud for tag

```python
# read data
df_main = pd.read_csv('./Merged_USvideos.csv')
print(df_main.shape)
df_main.head(3)
```

```python
df_main.describe()
```

```python
# convet all string to lower cases
all_tags = ' '.join(df_main['tags'].str.lower())
```

```python
stopwords = STOPWORDS
newStopWords = ['youtube']
stopwords.update(newStopWords)
tag_wordcloud = WordCloud(stopwords=stopwords, background_color="white", max_words=999999).generate(all_tags)
```

```python
rcParams['figure.figsize'] = 20, 30
plt.imshow(tag_wordcloud)
plt.axis("off")
plt.show()
```

```python
all_titles = ' '.join(df_main['title'].str.lower())
```

```python
stopwords = STOPWORDS
newStopWords = ['youtube']
stopwords.update(newStopWords)
title_wordcloud = WordCloud(stopwords=stopwords, background_color="white", max_words=999999).generate(all_titles)
```

```python
rcParams['figure.figsize'] = 20, 30
plt.imshow(title_wordcloud)
plt.axis("off")
plt.show()
```
